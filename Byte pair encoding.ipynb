```python
import numpy as np
from collections import defaultdict
import re

with open("corpus.txt", "r") as file:
    corpus = file.read().splitlines()
print("="*50)
print("Original Corpus")
print("="*50)

for word in corpus:
    print(word)
    
corpus = [word.lower().strip() for word in corpus]

print("\n")
print("="*50)
print("After Preprocessing")
print("="*50)

for word in corpus:
    print(word)
vocab = defaultdict(int)

for word in corpus:
    chars = " ".join(list(word)) + " </w>"
    vocab[chars] += 1
    
print("\n")
print("="*50)
print("Initial Vocabulary")
print("="*50)

for word, freq in vocab.items():
    print(word, ":", freq)
    
def get_pair_frequencies(vocab):
    pairs = defaultdict(int)
    
    for word, freq in vocab.items():
        symbols = word.split()
        
        for i in range(len(symbols)-1):
            pair = (symbols[i], symbols[i+1])
            pairs[pair] += freq
    return pairs
    
def merge_pair(pair, vocab):
    new_vocab = {}
    bigram = re.escape(" ".join(pair))
    replacement = "".join(pair)
    
    for word in vocab:
        new_word = re.sub(bigram, replacement, word)
        new_vocab[new_word] = vocab[word]
    return new_vocab
num_merges = 10

print("\n")
print("="*50)
print("BPE Merge Operations")
print("="*50)

for i in range(num_merges):
    pairs = get_pair_frequencies(vocab)
    if len(pairs) == 0:
        break
    print("\nPair Frequencies")
    
    for pair, freq in pairs.items():
        print(pair, ":", freq)
        
    best_pair = max(pairs, key=pairs.get)
    print("\nMost Frequent Pair :", best_pair)
    
    vocab = merge_pair(best_pair, vocab)
    print("\nVocabulary After Merge", i+1)
    
    for word, freq in vocab.items():
        print(word, ":", freq)
        
print("\n")
print("="*50)
print("FINAL VOCABULARY")
print("="*50)

for word, freq in vocab.items():
    print(word, ":", freq)
```
