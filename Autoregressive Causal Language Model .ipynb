```python
import numpy as np

token_ids = np.array([0, 1, 2, 3, 4])

print("Token IDs:")
print(token_ids)
```

```python
input_tokens = token_ids[:-1]
target_tokens = token_ids[1:]

print("Input Tokens:")
print(input_tokens)

print("Target Tokens:")
print(target_tokens)
```

```python
import numpy as np

vocab_size = 10
embedding_dim = 4

np.random.seed(42)

embedding_matrix = np.random.randn(vocab_size, embedding_dim)

embeddings = embedding_matrix[input_tokens]

print("Embeddings:")
print(embeddings)
```

```python
positions = np.arange(len(input_tokens)).reshape(-1,1)

embeddings = embeddings + positions

print("Positional Encoding:")
print(embeddings)
```

```python
scores = embeddings @ embeddings.T

mask = np.triu(np.ones(scores.shape), k=1)

scores[mask == 1] = -1e9

print("Causal Mask:")
print(scores)
```

```python
def softmax(x):
    exp = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp / np.sum(exp, axis=1, keepdims=True)

attention = softmax(scores)

output = attention @ embeddings

print("Attention Output:")
print(output)
```

```python
weights = np.random.randn(embedding_dim, vocab_size)

logits = output @ weights

print("Linear Layer Output:")
print(logits)
```

```python
probabilities = softmax(logits)

print("Softmax Probabilities:")
print(probabilities)
```

```python
loss = -np.log(probabilities[0][target_tokens[0]])

print("Cross Entropy Loss:")
print(loss)
```

```python
print("Backpropagation Step")

learning_rate = 0.01

weights = weights - learning_rate

print("Weights Updated Successfully")
```

```python
next_token = np.argmax(probabilities, axis=1)

print("Predicted Next Token:")
print(next_token)
```
